The closed form expression for $\left|{S_{21}} \right|$ when the monopole antenna is in the presence of a perfect conductor is $$ \left|S_{21}\right| = 2\left|\sin\left(\frac{\pi fd(1-\cos(2\theta)}{c\sin(\theta)}\right)\right|$$

Here we plot $\mid S_{21}\mid$ in dB for different values of $d$ and $\theta$ to observe the frequencies at which we get null values. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.arange(3e9, 7e9, 1.9e7)
d = [25, 30, 35, 40]
theta = [(np.pi)/2, (np.pi)/3]

for n in range(len(theta)):
    for i in range(len(d)):
        s = 20 * np.log10(2 * abs(np.sin(np.pi * d[i] * (1/1000) * f * (1 - np.cos(2 * theta[n])) * (1/3e8) * (1/np.sin(theta[n])))))
        
        title = r"$\theta$ = {} rad, d = {} mm "
        plt.plot(f, s)
        plt.title(title.format(round(theta[n],3), d[i]))
        plt.xlabel('Frequency (Hz)')
        plt.ylabel(r'$|S_{21}|$ (dB)')
        plt.ylim(-40, 10)
        plt.show()
```
