Here we plot the theoretical curves of $|S_{21}|$ in dB for different values of d, the distance from a monopole antenna to a perfect conductor in meters, to observe the frequencies at which we get null values. These frequencies correspond with the cusp seen on each of the plots.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.arange(3e9, 7e9, 1.9e7)
d = [.025, .030, .035, .040]

for i in range(len(d)):
    s = 20 * np.log10(2 * abs(np.sin(2 * np.pi * f * d[i] * (1/3e8))))
    plt.plot(f, s)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(r'$|S_{21}|$ (dB)')
    plt.show()
```
